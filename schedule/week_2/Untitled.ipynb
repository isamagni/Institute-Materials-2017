```python
from lxml import etree

with open('/Users/isamagni/Desktop/c026v-tok.xml', encoding='utf-8') as f:
    petrarch = etree.parse(f)

print("Got an ElementTree with root tag", petrarch.getroot().tag)
print(etree.tostring(petrarch).decode('utf-8'))
```

```python
def tei(tag):
    return "{http://www.tei-c.org/ns/1.0}%s" % tag

tei('text')
```

```python
for line in petrarch.iter(tei('l')):
    print(line.text)
```

```python
for line in petrarch.iter(tei('l')):
    content = line.text
    for child in line:
        if content is not None and child.tail is not None:
            content = content + child.tail
    print(content)
```
